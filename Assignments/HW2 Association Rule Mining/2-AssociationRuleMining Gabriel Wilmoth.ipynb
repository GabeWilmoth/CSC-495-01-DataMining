```python
import numpy as np    # for carrying out efficient computations
import pandas as pd   # for data manipulation and analysis
import seaborn as sns # library for making statistical graphics in Python
import matplotlib.pyplot as plt  #for visualizing the data
import math
import scipy 
from scipy import stats
import sklearn
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline  

from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
movies = pd.read_csv('../../AssignmentMaterials/Movies.tsv', sep='\t', header=0)
display(movies)
```

```python
ratings = pd.read_csv('../../AssignmentMaterials/Ratings.tsv', sep='\t', header=0)
display(ratings)
```

#### Merging movies and ratings dataframes on movieId

```python
full = pd.merge(movies, ratings, on='movieId')
display(full)
```

#### Filtering out movies with a rating of 4.0 or higher

```python
fullHighRating = full[~(full['rating'] < 4.0)] 
display(fullHighRating) 
```

```python
def listOfList(x):
    test = fullHighRating[fullHighRating['userId'] == x]
    lol = test['title'].values.tolist()
    testingp.append(lol)
```

```python
testingp = []
x = 1
while x <= 91:
    listOfList(x)
    x += 1
```

```python
print(testingp)
```

#### Using above Itemset to generate association rules via apriori

```python
te = TransactionEncoder()
te_ary = te.fit(testingp).transform(testingp)
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)

display(frequent_itemsets)
```

#### Generating the association rules and using a confidence metric of 0.8

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules["consequent_len"] = rules["consequents"].apply(lambda x: len(x))
rules[(rules['antecedent_len'] == 1) &
      (rules['consequent_len'] == 1)]
display(rules)
```

## Start of Genre Association Rule Mining

```python
display(fullHighRating)
```

```python
def listOfListGenre(x):
    test = fullHighRating[fullHighRating['userId'] == x]
    lol = test['genres'].values.tolist()
    genre.append(lol)
```

```python
genre = []
x = 1
while x <= 91:
    listOfListGenre(x)
    x += 1
```

```python
display(genre)
```

#### Using above Itemset to generate association rules via apriori

```python
te = TransactionEncoder()
te_ary = te.fit(genre).transform(genre)
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets_genre = apriori(df, min_support=0.4, use_colnames=True)

display(frequent_itemsets_genre)
```

#### Generating the association rules and using a confidence metric of 0.8

```python
rulesGenre = association_rules(frequent_itemsets_genre, metric="confidence", min_threshold=0.8)
rulesGenre["antecedent_len"] = rulesGenre["antecedents"].apply(lambda x: len(x))
rulesGenre["consequent_len"] = rulesGenre["consequents"].apply(lambda x: len(x))
rulesGenre[(rulesGenre['antecedent_len'] == 1) &
      (rulesGenre['consequent_len'] == 1)]
display(rulesGenre)
```

#### Interestingness Measures : Support & Confidence
    - Again I primarily chose to use Support and Confidence as we have more experience in said measures in class. I used a minimum support of 0.1 in the rating association and a minimum support of 0.4 in the genre association, in both ratings and genre I used a confidence of 0.8. With the ratings, there were few candidates produced with a minimum support greater than 0.1 and with both ratings and genre a confidence of 0.8 produced an ample amount of rules will at the same time not having too many rules to look at. Also, when we take a look at the confidence of ratings and genre rules we satisfy the case in which Confidence(X -> Y) > Support(Y) as otherwise, the results would be misleading.
